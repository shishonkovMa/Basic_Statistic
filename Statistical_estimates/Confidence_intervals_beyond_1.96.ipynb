# Confidence intervals: beyond 1.96


```python
import numpy as np
```

```python
import scipy.stats
```

```python
scipy.stats.norm.interval(0.95)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conf_level = 0.95
left, right = scipy.stats.norm.interval(0.95)
x = np.linspace(-7,7,200)
plt.plot(x, scipy.stats.norm.pdf(x))
x_int = np.linspace(left, right, 200)
plt.fill_between(x_int, scipy.stats.norm.pdf(x_int))
plt.plot(x, np.zeros_like(x))
print(left, right)
```

```python
conf_level = 0.9
left, right = scipy.stats.norm.interval(conf_level)
x = np.linspace(-7,7,200)
plt.plot(x, scipy.stats.norm.pdf(x))
x_int = np.linspace(left, right, 200)
plt.fill_between(x_int, scipy.stats.norm.pdf(x_int))
plt.plot(x, np.zeros_like(x))
print(left, right)
```

If we will increase a confidence level to 0.99

```python
conf_level = 0.99
left, right = scipy.stats.norm.interval(conf_level)
x = np.linspace(-7,7,200)
plt.plot(x, scipy.stats.norm.pdf(x))
x_int = np.linspace(left, right, 200)
plt.fill_between(x_int, scipy.stats.norm.pdf(x_int))
plt.plot(x, np.zeros_like(x))
print(left, right)
```

What if we do not have enough data to be sure that central limit theorem did it is job and if the distribution of our sample average then we divide it by the corresponding things that allow us to standartize it is not distributed according to normal law. We can use T-dictribution instead of using normal distribution.

```python
conf_level = 0.95
df = 5
left, right = scipy.stats.t.interval(conf_level, df=df)
x = np.linspace(-7,7,200)
plt.plot(x, scipy.stats.t.pdf(x, df=df))
x_int = np.linspace(left, right, 200)
plt.fill_between(x_int, scipy.stats.t.pdf(x_int, df=df))
plt.plot(x, np.zeros_like(x))
print(left, right)
```

## Example

sample_size = 10

```python
sample_size = 10
pop_mean = 3
pop_range = 4

catched = 0
iterations = 10000
for i in range(iterations):
    sample = np.random.uniform(pop_mean - pop_range, 
                               pop_mean + pop_range,
                               size=sample_size)
    stderr = np.std(sample, ddof=1) / np.sqrt(sample_size)
    left, right = scipy.stats.t.interval(conf_level, df=df)
    df = sample_size - 1
    r = right * stderr
    sample_mean = np.mean(sample)
    confint = (sample_mean - r, sample_mean + r)
    if confint[0] < pop_mean < confint[1]:
        catched += 1
        
catched / iterations
```

We can see that for very small samples and arbitrary distribution of our population in this case uniform distribution the t-distribution does not follow for actual distribution of the mean.

Properties of confidence intervals:
- then you increase the confidence level they became larger 
- if you decrease sample sizes if you increase amount of data they are also became larger 

Also we see that if we use t-distribution instead of normal distribution it makes our confident intervals more reliable than we have a little amount of data.

## Tasks

1. Consider a sample (1,2,2,4,3,2,5). Find 99% confidence interval for corresponding population mean using normal approximation.

```python
sample = [1,2,2,4,3,2,5]
conf_level = 0.99
sample_size = len(sample)

stderr = np.std(sample, ddof=1) / np.sqrt(sample_size)
left, right = scipy.stats.norm.interval(conf_level)
r = right * stderr
sample_mean = np.mean(sample)
confint = (sample_mean - r, sample_mean + r)
confint
```

2. F​or the same sample, find 99% confidence interval using t-distribution instead of normal distribution. (Thus we take into account that number of samples is rather small and normal approximation may be incorrect and produce too small confidence intervals.)

```python
df = sample_size - 1
left, right = scipy.stats.t.interval(conf_level, df=df)
r = right * stderr
sample_mean = np.mean(sample)
confint = (sample_mean - r, sample_mean + r)
confint
```
