# Two-sample t-test in Python

```python
from scipy.stats import ttest_ind
```

```python
sample1 = [3,4,5,4,6,7]
sample2 = [10,8,4,20,6,50,23,2]
```

Is it true that these two samples are generated by random variables with the same expected value or they genereted by random variables with different expected values?

```python
ttest_ind(sample1, sample2)#equal_var=False)
```

In this case (pvalue=0.13385879249866536) this test suggest you not to reject null hypotesis. We do not have enough evidence to conclude that sample2 came from a distribution with expected value that is different to their corresponding distribution from sample1.


**We can use two-tailed t-test with 'ttest_ind(sample1, sample2)'. Or we can use 'ttest_ind(sample1, sample2)'  pvalue divided by 2 for one-tailed t-test**
