# Task 1
Usually a farmer obtains 12 kg fruits per each tree. She is testing new experimental very expensive fertilizer and want to understand does it work or not. If trees produce more fruits with fertilizers, she will consider using it; otherwise (if it doesn't change productivity or decrease it) she will not use it. Five randomly selected trees with the new fertilizer produced 11 kg, 14 kg, 15 kg, 14 kg and 14 kg of fruits. Use t-test to find corresponding p-value.

```python
from scipy.stats import ttest_1samp, ttest_ind
import numpy as np
```

```python
mu_0 = 12
sample = [11, 14, 15, 14, 14]
stat, pvalue = ttest_1samp(sample, mu_0)
pvalue = pvalue / 2
pvalue
```

# Task 3
Find the value of t-statistics in two-sample t-test for samples: (2, 4, 6, 1, 3, 1, 4) and (3, 6, 2, 5, 7, 7).

```python
sample1 = [2, 4, 6, 1, 3, 1, 4]
sample2 = [3, 6, 2, 5, 7, 7]
t = ((np.mean(sample2) - np.mean(sample1)) / 
      np.sqrt(np.var(sample1, ddof=1)/len(sample1)+np.var(sample2, ddof=1)/len(sample2)))
t
```
