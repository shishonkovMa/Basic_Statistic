# Assignment
You perform a clinical trial to study new drug. You have 20 volunteers with some disease. You randomly split all the volunteers into two groups (10 volunteers in each): the treatment group and the control group. Volunteers in the treatment group receive the new drug, volunteers in the control group receive placebo (pills that looks like a drug but do not have active substance). You conclude that new drug is effective if people who take the drug will recover faster (on average) than people in the control group. If your drug is effective, you will invest in its production, otherwise you will look for another drug. Assume that you obtained the following data (disease duration in days).

## Questions

**1. Describe this problem in terms of statistical hypothesis testing framework. How would you model your data in terms of random variables?**

Answer: Is it true that these two samples are generated by random variables with the same expected value or they genereted by random variables with different expected values.

---

**2. State the null hypothesis and the alternative. Will your alternative be one-sided or two-sided? Why?**

Answer: `the alternative is one-sided`, since we need to determine the effectiveness of a new drug. And this will be revealed provided that the expected value of the treatment group is strictly less than the expected value of the control group ("strictly less", since an effective drug implies an acceleration of recovery, i.e. a reduction in the duration of the disease)

$$H_0: \mu_t = \mu_c$$
$$H_1: \mu_t < \mu_c$$ 
$$where\:t\:-\:treatment\:and\:c\:-\:control\:groups$$

---

**3. What kind of statistical test will you use? Why this test?**

Answer: we have two samples, on the data of which we must draw a conclusion about the equality or inequality of the two averages. We will not use a paired ttest here, because the samples is i.i.d. From this we conclude that we will use `two sample ttest`.

---

**4. Use this test (apply Python if necessary and provide your code), analyse the results and provide a conclusion in mathematical and real-life terms. Would you invest into production of this drug?**

```python
from scipy.stats import ttest_rel, ttest_ind


control_group = [6, 7, 7, 5, 7, 8, 8, 7, 7, 7]
treatment_group = [7, 6, 6, 5, 5, 6, 7, 5, 5, 8]

stat, pvalue = ttest_ind(treatment_group, control_group)
print('Statistic: ', stat, '\npvalue: ', pvalue / 2) # We have divided pvalue into 2, 
                                                     # since we have a one-sided alternative
```

We can see that `control_group` has a sample mean that is significantly larger than sample mean of `treatment_group`.<br>
Our pvalue is less than 5%.

Thus, we have evidence that the new drug is effective, since people taking the drug recover faster (on average) than people from the control group.
