```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sample_size = 10
pop_mean = 0.5
iterations = 10000
tvalues = []
for i in range(iterations):
    sample = np.random.uniform(0, 1, sample_size)
    t = ((np.mean(sample) - pop_mean) / 
         np.sqrt(np.var(sample, ddof=1)) * #ddof=1 - unbiased variance
         np.sqrt(sample_size))
    tvalues.append(t)
```

```python
sample
```

```python
plt.hist(tvalues, bins=30, density=True);
x = np.linspace(-7, 7, 200)
plt.plot(x, scipy.stats.norm.pdf(x))
plt.xlim(-7, 7)
```

```python
plt.hist(tvalues, bins=30, density=True);
x = np.linspace(-7, 7, 200)
plt.plot(x, scipy.stats.t.pdf(x, df=sample_size-1))
plt.xlim(-7, 7)
```

```python
plt.plot(x, scipy.stats.t.pdf(x, df=5), label='t, df=5')
plt.plot(x, scipy.stats.t.pdf(x, df=1), label='t, df=1')
plt.plot(x, scipy.stats.t.pdf(x, df=20), label='t, df=20')
plt.plot(x, scipy.stats.norm.pdf(x), label='norm')
plt.legend()
```
